```python
import sys
import math

pastX = []
pastY = []
good = []
landYG = 0
n = int(input())  
for i in range(n):
    land_x, land_y = [int(j) for j in input().split()]
    if not pastX:
        pastX.append(land_x)
        pastY.append(land_y)
    else:
        if land_y == pastY[-1]:
            good.append(pastX[-1])
            good.append(land_x)
            landYG = land_y
        else:
            pastX.pop()
            pastY.pop()
    
def dist(x:int, xm_d:int, xm_f:int )-> int:
    if xm_d < x < xm_f:
        distance = 0
    if x <= xm_d:
        distance = int(xm_d - x)
    if x >= xm_f:
        distance = int(x - xm_f)

    return distance

def reduce_horizontal_speed(hs):
    """
    Ajuste l'angle et la puissance pour réduire la vitesse horizontale (hs).
    """
    dr = 0  
    p = 0   

    if hs > 19:
        dr = 90  
        p = 4     
    elif hs < -19:
        dr = -90  
        p = 4    
    else:
        dr = 0  
        p = 0

    return dr, p


def reduce_vertical_speed(vs):
    """
    Ajuste la puissance pour réduire la vitesse verticale (vs).
    """
    dr = 0   
    p = 0   

    if vs < -39:
        p = 4 
    else:
        p = 0  
    return dr, p

dp = 0
dr:int = 0
time = 0

while True:
    x, y, hs, vs, f, r, p = [int(i) for i in input().split()]

    distance = dist(x, good[0], good[1])
    print(f"distance: {distance}", file=sys.stderr, flush=True)
    print(f"dp: {dp}", file=sys.stderr, flush=True)
    print(f"dp - distance: {dp - distance}", file=sys.stderr, flush=True)
    print(f"dp+60: {dp+60}", file=sys.stderr, flush=True)
    print(f"Hauteur: {y - landYG}", file=sys.stderr, flush=True)
    print(f"Y: {y}", file=sys.stderr, flush=True)
    print(f"HS: {hs}", file=sys.stderr, flush=True)
    print(f"VS: {vs}", file=sys.stderr, flush=True)

    hs_landing_limit = 27
    altitude_limit = 600
    hs_flight_limit = 39

    dr, p = 0, 0
    if distance == 0:
        # Aligner et stabiliser en fonction des vitesses actuelles
        dr_hs, p_hs = reduce_horizontal_speed(hs)
        dr_vs, p_vs = reduce_vertical_speed(vs)
        if (y - landYG) > altitude_limit:
            dr = dr_hs if dr_hs != 0 else dr_vs
            p = max(p_hs, p_vs)
        else:
            dr = 0
            p = 4


    elif x < good[0]:  # Si à gauche de la zone d'atterrissage
        if distance <= dp:  # Approche en cours
            if -hs_flight_limit < hs < hs_flight_limit:
                dr = 0  # Stabilisation
            else:
                dr = 65  # Corriger la trajectoire
            p = 4
        else:
            dr = -90  # Virage serré pour revenir
            p = 4

    elif x > good[1]:  # Si à droite de la zone d'atterrissage
        if distance <= dp:
            if -hs_flight_limit < hs < hs_flight_limit:
                dr = 0  # Stabilisation
            else:
                dr = -65  # Corriger la trajectoire
            p = 4
        else:
            dr = 90  # Virage serré pour revenir
            p = 4

    dp = distance

    # Commande finale pour le module
    print(f"{dr} {p}")
```

```python
import math
import random
from itertools import product

# --- Étape 1 : Simulateur de trajectoire ---
def simulate_trajectory(x, y, hs, vs, fuel, angle, thrust, landYG, max_steps=500):
    """
    Simule la trajectoire du module avec les paramètres donnés.
    """
    g = 3.711  # Gravité martienne
    trajectory = []

    for step in range(max_steps):
        if fuel <= 0 or y <= landYG:  # Stop si le carburant est épuisé ou si on touche le sol
            break

        # Calcul de l'accélération horizontale et verticale
        ax = -thrust * math.sin(math.radians(angle))
        ay = thrust * math.cos(math.radians(angle)) - g

        # Mise à jour des vitesses
        hs += ax
        vs += ay

        # Mise à jour des positions
        x += hs
        y += vs

        # Mise à jour du carburant
        fuel -= thrust

        # Ajouter l'état actuel à la trajectoire
        trajectory.append((x, y, hs, vs, fuel))

        # Arrêter si le lander dépasse la surface
        if y <= landYG:
            break

    return trajectory, (x, y, hs, vs, fuel)

# --- Étape 2 : Génération du terrain ---
def generate_terrain():
    """
    Génère un terrain martien aléatoire avec une zone d'atterrissage plate.
    """
    terrain = []
    last_x = 0
    last_y = random.randint(100, 1500)

    for _ in range(30):  # Génère 30 segments
        next_x = last_x + random.randint(100, 500)
        next_y = last_y + random.randint(-200, 200)
        terrain.append((next_x, next_y))
        last_x, last_y = next_x, next_y

    # Définir une zone plate pour l'atterrissage
    start_landing = random.randint(5, 20)
    end_landing = start_landing + 3
    landYG = terrain[start_landing][1]
    for i in range(start_landing, end_landing):
        terrain[i] = (terrain[i][0], landYG)

    return terrain, landYG

# --- Étape 3 : Génération de cas tests ---
def generate_test_cases(n_cases, terrain, landYG):
    """
    Génère des cas tests aléatoires basés sur le terrain fourni.
    """
    cases = []
    max_x = max([p[0] for p in terrain])
    for _ in range(n_cases):
        x = random.randint(0, max_x)
        y = random.randint(landYG + 500, landYG + 3000)
        hs = random.randint(-50, 50)
        vs = random.randint(-100, -10)
        fuel = random.randint(500, 1500)
        cases.append((x, y, hs, vs, fuel, landYG))
    return cases

# --- Étape 4 : Évaluation des poids ---
def evaluate_weights(cases, w_vs, w_hs, w_x, w_y, w_collision):
    """
    Évalue un ensemble de poids sur tous les cas tests.
    """
    total_score = 0
    for case in cases:
        x, y, hs, vs, fuel, landYG = case
        good = [0, 700]  # Zone d'atterrissage fictive
        peaks = []  # Pas d'obstacles ici pour simplifier

        _, final_state = simulate_trajectory(
            x, y, hs, vs, fuel, 0, 0, landYG
        )
        pred_x, pred_y, pred_hs, pred_vs, _ = final_state
        error = (
            abs(pred_vs) * w_vs +
            abs(pred_hs) * w_hs +
            abs(pred_x - (good[0] + good[1]) / 2) * w_x +
            abs(pred_y - landYG) * w_y +
            w_collision
        )
        total_score += error
    return total_score / len(cases)

# --- Étape 5 : Optimisation par Random Search ---
def random_search(cases, weight_ranges, n_trials=100):
    """
    Effectue un Random Search sur les poids de la fonction d'erreur.
    """
    best_score = float('inf')
    best_params = None

    for _ in range(n_trials):
        w_vs = random.randint(*weight_ranges['w_vs'])
        w_hs = random.randint(*weight_ranges['w_hs'])
        w_x = random.randint(*weight_ranges['w_x'])
        w_y = random.randint(*weight_ranges['w_y'])
        w_collision = random.randint(*weight_ranges['w_collision'])

        score = evaluate_weights(cases, w_vs, w_hs, w_x, w_y, w_collision)
        if score < best_score:
            best_score = score
            best_params = (w_vs, w_hs, w_x, w_y, w_collision)

    return best_params, best_score

# --- Main : Exécution ---
if __name__ == "__main__":
    # Générer un terrain
    terrain, landYG = generate_terrain()

    # Générer des cas tests
    cases = generate_test_cases(100, terrain, landYG)

    # Plages des poids pour la recherche
    weight_ranges = {
        "w_vs": (1, 10),
        "w_hs": (1, 10),
        "w_x": (1, 5),
        "w_y": (1, 5),
        "w_collision": (1000, 10000)
    }

    # Effectuer une recherche aléatoire sur 1000 essais
    best_params, best_score = random_search(cases, weight_ranges, n_trials=1000)

    # Afficher les résultats
    print(f"Meilleurs paramètres trouvés : {best_params}")
    print(f"Score associé : {best_score}")
```

```python
import sys
import math

pastX = []
pastY = []
good = []
landYG = 0
target_center = 0

n = int(input())  
for i in range(n):
    land_x, land_y = [int(j) for j in input().split()]
    # print(f"land_y: {land_y}", file=sys.stderr, flush=True)
    if not pastX:
        pastX.append(land_x)
        pastY.append(land_y)
    else:
        if land_y == pastY[-1]:
            good.append(pastX[-1])
            good.append(land_x)
            landYG = land_y
        pastX.append(land_x)
        pastY.append(land_y)

sommet = [el for el in pastY if el > landYG]
def find_peaks(pastX, pastY, landYG):
    """
    Trouve les coordonnées des sommets (x, y) où y > landYG.
    """
    peaks = [(x, y) for x, y in zip(pastX, pastY) if y > landYG]
    return peaks

peaks = find_peaks(pastX,pastY,landYG)

def check_trajectory_against_peaks(trajectory, peaks):
    """
    Vérifie si la trajectoire touche ou passe sous les sommets.
    Retourne True si une collision ou un passage sous un sommet est détecté.
    """
    for x, y, _, _ in trajectory:
        for px, py in peaks:
            if abs(x - px) <= 10 and y < py: 
                return True
    return False

print(f"good: {good}", file=sys.stderr, flush=True)
target_center = good[0] + ((good[1]-good[0])//2)

def print_ctrl(distance, distanceT, dp, y, hs, vs, landYG):
    print(f"target center: {target_center}", file=sys.stderr, flush=True)
    print(f"distance: {distance}", file=sys.stderr, flush=True)
    print(f"distance du centre: {distanceT}", file=sys.stderr, flush=True)
    print(f"dp: {dp}", file=sys.stderr, flush=True)
    print(f"dp - distance: {dp - distance}", file=sys.stderr, flush=True)
    print(f"dp+60: {dp+60}", file=sys.stderr, flush=True)
    print(f"Hauteur: {y - landYG}", file=sys.stderr, flush=True)
    print(f"Y: {y}", file=sys.stderr, flush=True)
    print(f"landYG: {landYG}", file=sys.stderr, flush=True)
    print(f"VS: {vs}", file=sys.stderr, flush=True)
    
    return 0

def dist_center(x, xc):
    if x <= xc:
        distance = int(xc - x)
    else:
        distance = int(x - xc)
    return distance

def dist(x:int, xm_d:int, xm_f:int )-> int:
    if xm_d < x < xm_f:
        distance = 0
    if x <= xm_d:
        distance = int(xm_d - x)
    if x >= xm_f:
        distance = int(x - xm_f)
    return distance

def simulate_trajectory(x, y, hs, vs, f, angle, thrust, landYG, peaks, max_steps=500):
    """
    Simule la trajectoire du module et renvoie la vitesse et position finale.
    """
    g = 3.711  # Gravité martienne
    trajectory = []
    
    for step in range(max_steps):
        # Calcul de l'accélération horizontale et verticale
        ax = -thrust * math.sin(math.radians(angle))
        ay = thrust * math.cos(math.radians(angle)) - g

        # Mise à jour des vitesses
        hs += ax
        vs += ay

        # Mise à jour des positions
        x += hs
        y += vs

        # Trajectoire
        trajectory.append((x, y, hs, vs))

        if thrust > 0:
            f -= thrust
            if f <= 0: 
                thrust = 0

        # Arrêter si on atteint ou dépasse le sol
        if y <= landYG+10:
            break
    touches_peak = check_trajectory_against_peaks(trajectory, peaks)
    # Retourne la trajectoire complète et l'état final
    return trajectory, (x, y, hs, vs), touches_peak

def optimize_landing(x, y, hs, vs, f, landYG, good, peaks, distance, max_steps=500):
    """
    Optimise les commandes pour un atterrissage vertical et sécurisé.
    """
    print(f"Opti phase", file=sys.stderr, flush=True)
    g = 3.711  # Gravité martienne
    best_angle, best_thrust = 0, 0
    min_error = float('inf')
    best_final_state = None

    coef_1 = 4
    coef_2 = 6

    if y - landYG > 1000:  # Phase haute
        angle_step = 1
    else:  # Phase basse
        angle_step = 1

    for angle in range(-90, 91, angle_step):  # Explore les angles par pas de 1°
        for thrust in range(0, 5):  # Explore les poussées possibles
            _, (pred_x, pred_y, pred_hs, pred_vs), touches_peak = simulate_trajectory(
                x, y, hs, vs, f, angle, thrust, landYG, peaks, max_steps
            )
            sommetE = float('inf') if touches_peak else 0
            # Vérifie si la position finale est dans la zone cible
            if good[0] <= pred_x <= good[1]: #and pred_y <= landYG:
                # Calcul de l'erreur pour évaluer la trajectoire
                # Penalise les vitesses non sûres et favorise la verticalité
                error = (
                    (
                    abs(pred_vs) * coef_1 +  
                    abs(pred_hs) * coef_2
                    ) +
                    abs(pred_x - (good[0] + good[1]) / 2) +
                    abs(pred_y - landYG)*2 +
                    sommetE
                )
                
                if error < min_error:
                    min_error = error
                    best_angle, best_thrust = angle, thrust
                    best_final_state = (pred_x, pred_y, pred_hs, pred_vs)


    return best_angle, best_thrust, best_final_state

dp = 0
dr:int = 0
time = 0

while True:
    x, y, hs, vs, f, r, p = [int(i) for i in input().split()]

    distance = dist(x, good[0], good[1])
    distanceT = dist_center(x, target_center)
    direction = -1 if x <target_center else 1
    
    angle = dr  # Angle actuel
    thrust = p  # Poussée actuelle
    trajectory, final_state, touches_peak = simulate_trajectory(x, y, hs, vs, f, angle, thrust, landYG, peaks)
    pred_x, pred_y, pred_hs, pred_vs = final_state
    
    print_ctrl(distance, distanceT, dp, y, hs, vs, landYG)
    
    print(f"------------------------------------", file=sys.stderr, flush=True)
    print(f"touches_peak: =>{touches_peak}", file=sys.stderr, flush=True)
    print(f"Prévision de position finale: x={pred_x}, y={pred_y}", file=sys.stderr, flush=True)
    print(f"Vitesse horizontale finale: hs={pred_hs}", file=sys.stderr, flush=True)
    print(f"Vitesse verticale finale: vs={pred_vs}", file=sys.stderr, flush=True)

    hs_landing_limit = 25
    altitude_limit = 300
    hs_flight_limit = 50
        
    dp = distance
    time +=1

    dr, p, final_state = optimize_landing(x, y, hs, vs, f, landYG, good, peaks, distance)
    print(f"final_state ---->{final_state}", file=sys.stderr, flush=True)

    # Phase finale stricte
    if y - landYG < 65 and distance == 0:  # Proche du sol
        print(f"final S", file=sys.stderr, flush=True)
        dr = 0
        p = 4 
    
    print(f'{dr} {p}')
```
